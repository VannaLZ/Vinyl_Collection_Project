```python
collection_sum = df_higher.groupby('Collection')['Higher_230703'].sum().reset_index().sort_values('Higher_230703')
collection_sum
```

```python
collection_avg = df_higher.groupby('Collection')['Higher_230703'].mean().reset_index().sort_values('Higher_230703')
collection_avg = collection_avg.rename(columns={'Higher_230703': 'Average_230703'})
collection_avg['Average_230703'] = collection_avg['Average_230703'].round(2)
collection_avg
```

```python
columns_to_exclude = ['Artist', 'Title']  # Specify the columns to exclude from calculations
collection_avg = pd.DataFrame()  # Create an empty DataFrame to store the results

for column in df_higher.columns:
    if column not in columns_to_exclude and column != 'Collection':  # Exclude specified columns
        column_avg = df_higher.groupby('Collection')[column].mean().reset_index()
        column_avg = column_avg.rename(columns={column: f'Average_{column}'})
        column_avg[f'Average_{column}'] = column_avg[f'Average_{column}'].round(2)
        collection_avg = pd.concat([collection_avg, column_avg[f'Average_{column}']], axis=1)

collection_avg['Collection'] = df_higher['Collection'].unique()  # Add the 'Collection' column to the DataFrame
```
