I  create three separate DataFrames (df_min, df_median, and df_higher) are created by indexing the original DataFrame (df) with the corresponding column lists.

Each containing the columns with 'Min', 'Median', and 'Higher' respectively.

```python
min_columns = [col for col in df.columns if 'Min' in col]
median_columns = [col for col in df.columns if 'Median' in col]
higher_columns = [col for col in df.columns if 'Higher' in col]

df_min = df[['Artist', 'Title', 'Collection'] + min_columns]
df_median = df[['Artist', 'Title', 'Collection'] + median_columns]
df_higher = df[['Artist', 'Title', 'Collection'] + higher_columns]
```

```python
#check the 10 higher values taking in conosideration the last column which is the last entry.
df_min_sorted = df_min.sort_values("Min_230703", ascending=False)
df_min_sorted.head(10)
```

```python
#check the 10 higher values taking in conosideration the last column which is the last entry.
df_median_sorted = df_median.sort_values("Median_230703", ascending=False)
df_median_sorted.head(10)
```

```python
#check the 10 higher values taking in conosideration the last column which is the last entry.
df_higher_sorted = df_higher.sort_values("Higher_230703", ascending=False)
df_higher_sorted.head(10)
```

```python
df_higher.loc[:, 'Difference'] = df_higher.iloc[:, -2] - df_higher.iloc[:, 4]
```

```python
positive_difference = df_higher[df_higher['Difference'] > 0]
positive_higher_difference = positive_difference.sort_values("Difference", ascending=False)
```
