```python
import re

# define the regular expression pattern to extract the values
pattern = r"Lowest:\s*([^\n]+)\nMedian:\s*([^\n]+)\nHighest:\s*([^\n]+)"

# extract the values and create new columns
df[["Lowest", "Median", "Highest"]] = df["Notes"].str.extract(pattern)

# remove the "t" character from the values
df["Lowest"] = df["Lowest"].str.replace("t", "")
df["Median"] = df["Median"].str.replace("t", "")
df["Highest"] = df["Highest"].str.replace("t", "")
```
