```python
import sys
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
client_id = os.getenv("client_id")
client_secret = os.getenv("client_secret")
credentials = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(client_credentials_manager=credentials)
```

```python
table = pd.read_csv('1990_LPs.csv')
```

```python
table.sample()
```

```python
for index, row in table.iterrows():
    for i in range(1, 6):
        artist_id = row[f'artist_id_{i}']
        if artist_id and artist_id != "none":
            artist = sp.artist(artist_id)
            genres = artist['genres']
            table.at[index, f'genre_{i}'] = ", ".join(genres)
        else:
            table.at[index, f'genre_{i}'] = "none"
```

```python
pd.set_option('display.max_columns', None)
```

```python
table.sample(5)
```

```python
# Get a list of the column names in the current order
current_cols = table.columns.tolist()
print(current_cols)
```

```python
# Define the new order for the columns
new_cols = ['song', 'artist_1', 'artist_id_1', 'genre_1', 'album', 'label', 'popularity', 'total_tracks', 'album_id',
            'duration','date', 'date_album', 'imagen', 'url_song', 'artist_2', 'genre_2','artist_id_2',
            'artist_3','genre_3','artist_id_3', 'artist_4', 'artist_id_4',  'genre_4', 'artist_5', 'artist_id_5',
             'genre_5']

# Reorder the columns in the table
table = table.reindex(columns=new_cols)
```

```python
table.sample()
```

```python
# Save the updated table to a new file
table.to_csv('1990_LPs.csv', index=False)
```
